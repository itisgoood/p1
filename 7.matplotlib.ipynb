```python
import matplotlib.pyplot as plt
```

```python
# 스타일 적용하기
plt.style.use('classic')
```

```python
import numpy as np
```

```python
#샘플 데이터 생성 넘파이에서 제공하는걸로 만들어볼꺼임
x = np.linspace(0,10,100)
x
```

```python
plt.plot(x, np.sin(x)) #싸인그래프
```

```python
plt.plot(x,np.cos(x)) #코싸인 그래프
```

```python
#그래프 두개를 한번에 그리고 싶다
plt.plot(x, np.sin(x))
plt.plot(x,np.cos(x))
```

```python
fig= plt.figure() #그림판을 만드는것
plt.plot(x, np.sin(x),'-')
plt.plot(x,np.cos(x),'--') # -, --표현의 방식
fig.savefig('my_figure.png') #파일을 만들고 (옆에 라인) 저장
```

```python
fig.canvas.get_supported_filetypes() #파일의 확장자들 어떤걸 사용해야 더 명확하게 보이나
```

```python
fig=plt.figure()
ax = plt.axes() 

x = np.linspace(0,10,100)
ax.plot(x,np.sin(x))
```

```python
plt.plot (x, np.sin(x-0), color='blue')
plt.plot (x, np.sin(x-1), color='g')
plt.plot (x, np.sin(x-2), color='0.75') #0과 1 사이의 회색조를 표현하는 색깔 :1이면 흰색 0이면 검은색
plt.plot (x, np.sin(x-3), color='#FFDD44')#16진수로 표현한 RGB칼라
plt.plot (x, np.sin(x-4), color=(1.0,0.2,0.3)) #10진수로 RGB값을 튜플로표현

```

```python
plt.plot(x,x+0, linestyle = 'solid')
plt.plot(x,x+1, linestyle = 'dashed')
plt.plot(x,x+2, linestyle = 'dashdot')
plt.plot(x,x+0, linestyle = 'dotted')

plt.plot(x,x+4, linestyle = '-')
plt.plot(x,x+5, linestyle = '--')
plt.plot(x,x+6, linestyle = '-.')
plt.plot(x,x+7, linestyle = ':')
```

```python
plt.plot(x,x+0,'-g')#선스타일하고 색을 한번에 지정하는것 
plt.plot(x,x+1,'--c')
plt.plot(x,x+2,'-r')
plt.plot(x,x+3,'-,k')
plt.plot(x,x+4,':r')
```

```python
#플롯 조정하기 : 축 경계
plt.plot(x,np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axlim('tight')
```

```python
#플롯에 레이블 붙이기
plt.plot(x, np.sin(x))
plt.title('A SineCurve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.axis('equal')
plt.legend()
```

```python
#산점도의 출력 (점으로 표현하는 것)
x = np.linspace(0,10,30)
y= np.sin(x)
plt. plot(x,y,'o',color='black') # o 자리는 마커를 사용하는 자리
```

```python
#마커 종류별로 사용
rng = np.random.RandomState(0)
for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5),rng.rand(5),marker,label='marker={0}'.format(marker))
```

```python
#마커 종류별로 사용
rng = np.random.RandomState(0)
for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5),rng.rand(5),marker,label='marker={0}'.format(marker))
  plt.legend() #표시
  plt.xlim(0,1.8) # 범위설정
```
