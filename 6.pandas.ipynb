```python
#기초분석을 판다스로 사용
import pandas as pd # 판다스 불러올께 근데 이름 너무 길어서 pd로 할께
```

```python
german = pd.read_csv('http://freakonometrics.free.fr/german_credit.csv')
```

```python
type(german)
```

```python
german.columns.values
```

```python
german_sample = german[['Creditability','Duration of Credit (month)','Purpose','Credit Amount']]
```

```python
german_sample
```

```python
german_sample.min()
```

```python
german_sample.max()
```

```python
german_sample.mean()
```

```python
german_sample.head() #head()T쓰면 앞부분의 데이터 5개를 맛보기로 보여줌
```

```python
german_sample.corr()#판다스 상관계수분석
```

```python
#Group By를 이용한 계산 및 요약 통계
german_sample = german[['Credit Amount', 'Type of apartment']]
german_sample
```

```python
german_grouped = german_sample['Credit Amount'].groupby(german_sample['Type of apartment'])
```

```python
german_grouped.mean()
```

```python
german_grouped.max()
```

```python
german_sample = german[['Type of apartment','Sex & Marital Status','Credit Amount']]
```

```python
for type,group in german_sample.groupby('Type of apartment'):
  print(type)
  print(group.head())
```

```python
for (type, sex), group in german_sample.groupby('Type of apartment','Sex & Maritla Status'):
  print(type,sex)
  print(group.head(n=3))
```

```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
```

```python
births.shape
```

```python
births.head
```

```python
births['year']//10 #나눠주고
```

```python
births['year']//10*10 # 년대가 나옴
```

```python
births['decade']=births['year'] #decade에 내용을 저장
```

```python
births.head()
```

```python
births.pivot_table('births',index='decade', columns = 'gender',aggfunc='sum')
```
